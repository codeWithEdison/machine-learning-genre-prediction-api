```python
import pandas as pd

data = pd.read_csv('music.csv')
data
```

```python
import matplotlib.pyplot as plt
## histogram 
plt.hist(data['age'], color='skyblue', edgecolor ='black')
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('histogram of age in music data')#
```

```python
plt.plot(data['age'])
```

```python
#scatter
plt.scatter(data['age'] ,data['genre'] )
plt.xlabel('Age')
plt.ylabel('Genre')
plt.title('Age vs Genre')
```

```python
data['genre'].value_counts().plot(kind='bar')
```

```python
import seaborn as sns
correlation = data.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation Heatmap')
```

```python
#box plot
sns.boxplot(x='gender', y='age', data=data)
plt.xlabel('gender')
plt.ylabel('Age')
plt.title('Age Distribution by Genre')
```
